# Feature Engineering with Categorial Data

```python
import numpy as np
import pandas as pd

category_ids = list(range(4))
category_labels = ['CategoryA', 'CategoryB', 'CategoryC', 'CategoryD']
category_mapping = pd.DataFrame(list(zip(category_ids, category_labels)), columns=['Id', 'Name'])
category_mapping
```

```python
df = pd.read_csv("./data/time-data.csv", delimiter=";", header=None)
df[0] = df[0].apply(pd.to_datetime)
df.head()
```

```python
# OneHotEncoder for multiple categorial columns
# from sklearn.preprocessing import OneHotEncoder
# LabelBinarizer for binarizing one categorial column
from sklearn.preprocessing import LabelBinarizer

# Encode category
lb = LabelBinarizer()
binarized_labels = lb.fit_transform(df[3])
df[["Category" + str(i) for i in range(binarized_labels.shape[1])]] = binarized_labels
df.head()
```
